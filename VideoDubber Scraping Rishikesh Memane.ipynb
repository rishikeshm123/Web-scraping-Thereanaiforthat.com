```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import pandas as pd
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import selenium
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
email = "Enter Email"
password = "Enter Password"
```

```python
urls = ["https://theresanaiforthat.com/saved/"]

## Opening Chrome Window   ################################## 

## change chrome driver path

s = Service(r"C:\Users\Rishi\Downloads\chromedriver\chromedriver.exe")
for url in urls:
    driver = webdriver.Chrome(service=s)
    driver.maximize_window()
    driver.get(url)
```

```python
username = driver.find_element(By.XPATH,'//input[@class="whsOnd zHQkBf"]')
username.click()
username.send_keys(email)
username.send_keys(Keys.RETURN)
```

```python
password_element = driver.find_element(By.XPATH,'//input[@class="whsOnd zHQkBf"]')
password_element.click()
password_element.send_keys(password)
password_element.send_keys(Keys.RETURN)
```

```python
# Extracting each element and storing the text in a list

elements = driver.find_elements(By.XPATH,'//li[@class="li m"]')
element_text = []
for i in elements:
    element_text.append(i.text)

element_text = element_text[3:]
```

```python
# Converting list into a dataframe

df = pd.DataFrame({"Data":element_text})
```

```python
# replacing '\n' by commas and splitting dataframe on commas
df['Data'] = df['Data'].str.replace('\n',',')
df_final = df['Data'].str.split(',', expand = True)

# indexing only required columns and renaming

df_final = df_final.iloc[:, :3]
df_final.rename(columns = {0 : 'Names' , 1: 'Category', 2:'Saves'}, inplace = True) 

# Changing datatye of number of saves to integer and sort

df_final['Saves'] = pd.to_numeric(df_final['Saves'], errors='coerce').astype('Int64')
df_final.fillna(0,inplace = True)
df_final.sort_values(by = 'Saves', ascending =False)
```
